```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('./Cellphone.csv')
print(df.size)
df.head()
```

```python
df.columns[df.isnull().any()]
```

```python
corr_matrix = df.corr()

# Create a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

#### Based on the above graph, there is clear correlation between 1) price and ppi and 2) price and ram

```python
graph_vars = [["ppi","Price"],["ram", "Price"]]
for i, (x, y) in enumerate(graph_vars):
    plt.figure(i)
    plt.scatter(df[x], df[y])
    plt.xlabel(x)
    plt.ylabel(y)
    plt.title(f"Scatter plot of {x} vs {y}")

plt.show()
```

### Linear regression using OLS

```python
from sklearn.linear_model import LinearRegression

# Example data
X = df[["ppi"]]
y = df[["Price"]]

# Create and fit the model
model = LinearRegression()
model.fit(X, y)

# Get the parameters
theta_ols = [model.intercept_, model.coef_[0]]
print("Parameters using OLS:", theta_ols)
```
