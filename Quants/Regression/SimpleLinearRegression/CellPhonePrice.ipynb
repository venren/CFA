```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('./Cellphone.csv')
print(df.size)
df.head()
```

```python
df.columns[df.isnull().any()]
```

```python
corr_matrix = df.corr()

# Create a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

#### Based on the above graph, there is clear correlation between 1) price and ppi and 2) price and ram

```python
graph_vars = [["ppi","Price"],["ram", "Price"]]
for i, (x, y) in enumerate(graph_vars):
    plt.figure(i)
    plt.scatter(df[x], df[y])
    plt.xlabel(x)
    plt.ylabel(y)
    plt.title(f"Scatter plot of {x} vs {y}")

plt.show()
```

### Linear regression using OLS

```python
from sklearn.linear_model import LinearRegression

# Example data
X = df[["ppi"]]
y = df[["Price"]]

# Create and fit the model
model = LinearRegression()
model.fit(X, y)

# Get the parameters
theta_ols = [model.intercept_, model.coef_[0]]
print("Parameters using OLS:", theta_ols)

# Predictions
y_pred = model.predict(X)

# Plot data points
plt.scatter(X, y, color='blue', label='Data Points')

# Plot regression line
plt.plot(X, y_pred, color='red', label='Regression Line')

# Add labels and legend
plt.xlabel('ppi')
plt.ylabel('Price')
plt.grid(True)
plt.legend()

# Show plot
plt.show()

# Plot residuals with independent variable (X in this case)
residuals = y - y_pred

# Plot residuals against predicted values
plt.scatter(X, residuals, color='blue')
plt.axhline(y=0, color='red', linestyle='--')
plt.title('OLS Linear Regression')
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.grid(True)
plt.show()
```

#### Based on above, looks like value of residuals increases as number of PPI increases. 
#### So, beyound certain PPI value, there are additional factors that should be considered to decide the price of mobile. 

```python
from sklearn.linear_model import SGDRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
import numpy as np

# Sample data
X = df[["ppi"]]
y = df[["Price"]]

# Create a pipeline with feature scaling and linear regression using SGD
model = make_pipeline(StandardScaler(), SGDRegressor(max_iter=1000, tol=1e-3))

# Fit the model
model.fit(X, y)

# Print learned parameters
print("Learned weights:", model.named_steps['sgdregressor'].coef_)
print("Learned bias:", model.named_steps['sgdregressor'].intercept_)

# Plot the data points
plt.scatter(X, y, color='blue', label='Data Points')

# Plot the line predicted by the model
X_range = np.linspace(np.min(X), np.max(X), len(X)).reshape(-1, 1)
y_pred = model.predict(X_range)
plt.plot(X_range, y_pred, color='red', label='SGD Model')

plt.xlabel('PPI')
plt.ylabel('Price')
plt.title('SGD Linear Regression')
plt.legend()
plt.grid(True)
plt.show()

# Plot residuals with independent variable (X in this case)
residuals = y.to_numpy().flatten() - y_pred

# Plot residuals against predicted values
plt.scatter(X.to_numpy().flatten(), residuals, color='blue')
plt.axhline(y=0, color='red', linestyle='--')
plt.title('SGD Linear Regression')
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.grid(True)
plt.show()
```
